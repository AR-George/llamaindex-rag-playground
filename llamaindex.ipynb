```python
from llama_index.core import SimpleDirectoryReader

documents = SimpleDirectoryReader("./data").load_data()
```

```python
from azure.identity import DefaultAzureCredential

aoai_token_scope = "https://cognitiveservices.azure.com/.default"
credential = DefaultAzureCredential()
token_response = credential.get_token(aoai_token_scope)
aoai_key = token_response.token
```

```python
import os
from llama_index.llms.azure_openai import AzureOpenAI
from llama_index.embeddings.azure_openai import AzureOpenAIEmbedding

aoai_endpoint = os.getenv('AZURE_OPENAI_ENDPOINT')
aoai_api_version = os.getenv('OPENAI_API_VERSION')

llm = AzureOpenAI(
    deployment_name=os.getenv('CHAT_MODEL_NAME'),
    azure_endpoint=aoai_endpoint,
    api_key=aoai_key,
    api_version=aoai_api_version,
)

embed_model = AzureOpenAIEmbedding(
    deployment_name=os.getenv('EMBEDDING_MODEL_NAME'),
    azure_endpoint=aoai_endpoint,
    api_key=aoai_key,
    api_version=aoai_api_version,
)
```

```python
# global
from llama_index.core import Settings
Settings.llm = llm
Settings.embed_model = embed_model
```

```python
from llama_index.core import VectorStoreIndex

# index
vector_index = VectorStoreIndex.from_documents(documents)
query_engine = vector_index.as_query_engine()

response = query_engine.query("Cats spend a significant portion of their lives grooming themselves")
print(response)
```
